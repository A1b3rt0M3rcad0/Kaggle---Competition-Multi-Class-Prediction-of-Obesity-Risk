# Importando bibliotecas

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from skopt import dummy_minimize

import pandas as pd
import numpy as np
import pickle as pkl
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
%matplotlib inline
```

# Carregando dados

```python
df = pd.read_csv('dataset/train.csv')
df.head()
```

```python
df.index = df['id']
df.drop(columns=['id'], inplace=True)
```

```python
df.info()
```

# Encode dos dados categóricos

```python
test_dataset = pd.read_csv('dataset/test.csv')
```

```python
__temp_df = pd.concat([test_dataset, df], axis=0)
__temp_df.drop(columns=['NObeyesdad', 'id'], inplace=True)
```

```python
columns_to_encode = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS', 'NObeyesdad']
encoders = {}
for column in columns_to_encode:
    if column == 'NObeyesdad':
        le = LabelEncoder()
        le.fit(df['NObeyesdad'])
        encoders['NObeyesdad'] = le
        continue
    le = LabelEncoder()
    le.fit(__temp_df[column])
    __temp_df[column] = le.transform(__temp_df[column])
    encoders[column] = le
```

```python
for k, encoder in encoders.items():
    df[k] = encoder.transform(df[k])
```

```python
# Salvando Ecoders
with open('encoders.pkl', 'wb') as f:
    pkl.dump(encoders, f)
```

# Dividindo dados entre previsores e labels

```python
y = df['NObeyesdad']
X = df.drop(columns=['NObeyesdad'])
```

# Selecionando o modelo com melhor desempenho

```python
models = {
    'RandomForestClassifier': RandomForestClassifier(n_jobs=-1),
    'MLPClassifier': MLPClassifier(solver='adam', max_iter=400),
    'SVC': SVC(max_iter=200),
    'LogisticRegression': LogisticRegression(n_jobs=-1, max_iter=200),
    'GaussianNB': GaussianNB(),
    'LGBMClassifier': LGBMClassifier(),
    'XGBClassifier': XGBClassifier()
}
scoring = {}

for name, model in models.items():
    skf = StratifiedKFold(n_splits=10)
    scoring[name] = {
        'accuracy_score': [],
        'confusion_matrix': []
    }
    for i, (train_index, test_index) in enumerate(skf.split(X, y)):
        model.fit(X.iloc[train_index], y.iloc[train_index])
        predict = model.predict(X.iloc[test_index])
        acc = accuracy_score(y.iloc[test_index], predict)
        cm = confusion_matrix(y.iloc[test_index], predict)
        scoring[name]['accuracy_score'].append(acc)
        scoring[name]['confusion_matrix'].append(cm)
```

```python
mean_score_metrics_per_model = {   
}

for k, metrics in scoring.items():
    mean_score_metrics_per_model[k] = {
        'mean_accuracy_score': [np.mean(metrics['accuracy_score']), np.std(metrics['accuracy_score'])],
        'mean_cm': sum(metrics['confusion_matrix']) / len(metrics['confusion_matrix'])
    }
```

```python
ax = sns.heatmap(mean_score_metrics_per_model['RandomForestClassifier']['mean_cm'], annot=True)
ax.set_title('RandomForestClassifier Confusion Matrix');
mean_score_metrics_per_model['RandomForestClassifier']['mean_accuracy_score']
```

```python
ax = sns.heatmap(mean_score_metrics_per_model['MLPClassifier']['mean_cm'], annot=True)
ax.set_title('MLPClassifier Confusion Matrix');
mean_score_metrics_per_model['MLPClassifier']['mean_accuracy_score']
```

```python
ax = sns.heatmap(mean_score_metrics_per_model['SVC']['mean_cm'], annot=True)
ax.set_title('SVC Confusion Matrix');
mean_score_metrics_per_model['SVC']['mean_accuracy_score']
```

```python
ax = sns.heatmap(mean_score_metrics_per_model['LogisticRegression']['mean_cm'], annot=True)
ax.set_title('LogisticRegression Confusion Matrix');
mean_score_metrics_per_model['LogisticRegression']['mean_accuracy_score']
```

```python
ax = sns.heatmap(mean_score_metrics_per_model['GaussianNB']['mean_cm'], annot=True)
ax.set_title('GaussianNB Confusion Matrix');
mean_score_metrics_per_model['GaussianNB']['mean_accuracy_score']
```

```python
ax = sns.heatmap(mean_score_metrics_per_model['LGBMClassifier']['mean_cm'], annot=True)
ax.set_title('LGBMClassifier Confusion Matrix');
mean_score_metrics_per_model['LGBMClassifier']['mean_accuracy_score']
```

```python
ax = sns.heatmap(mean_score_metrics_per_model['XGBClassifier']['mean_cm'], annot=True)
ax.set_title('XGBClassifier Confusion Matrix');
mean_score_metrics_per_model['XGBClassifier']['mean_accuracy_score']
```

# Selecionando as melhores features

#### Usando XGBOOST

```python
feature_score = {i: [] for i in range(1, X.shape[1] + 1)}

for i in range(1, X.shape[1] + 1):
    skl = SelectKBest(chi2, k=i)
    X_filtered = skl.fit_transform(X, y)
    skf = StratifiedKFold(n_splits=10)
    _temp_scoring = {
        'accuracy_score': [],
        'confusion_matrix': []
    }
    for train_index, test_index in skf.split(X, y):
        xgb = XGBClassifier()
        xgb.fit(X_filtered[train_index], y[train_index])
        predicts = xgb.predict(X_filtered[test_index])
        acc = accuracy_score(y[test_index], predicts)
        cm = confusion_matrix(y[test_index], predicts)
        _temp_scoring['accuracy_score'].append(acc)
        _temp_scoring['confusion_matrix'].append(cm)
    _temp_scoring['accuracy_score'] = np.mean(_temp_scoring['accuracy_score'])
    _temp_scoring['confusion_matrix'] = sum(_temp_scoring['confusion_matrix']) / len(
        _temp_scoring['confusion_matrix']
    )
    feature_score[i].append(_temp_scoring)
```
